```python
import os

# Get the current working directory
current_directory = os.getcwd()

print("Current working directory:", current_directory)
```

```python
import pandas as pd

restos_2 = pd.read_csv("/Users/sarakim/Downloads/restos_2_mar_24.csv")

# Display the first few rows of the DataFrame to verify it was read correctly
print(restos_2.head())
```

```python
# Filter rows where the 'latitude' column has a value of "0"
zero_latitude_rows = restos_2[restos_2['latitude'] == 0]

# Display the first 20 rows of the 'latitude' column for rows with latitude equal to "0"
first_20_zero_latitude = zero_latitude_rows['latitude'].head(20)

print("First 20 rows of 'latitude' column where value equals '0':")
print(first_20_zero_latitude)
```

```python
# Count the number of rows where the 'latitude' column has a value of "0"
num_zero_latitude_rows = len(restos_2[restos_2['latitude'] == 0])

print("Number of rows with a '0' value in the 'latitude' column:", num_zero_latitude_rows)
```

```python
# Filter rows where both "latitude" and "longitude" are not 0
restos_3 = restos_2[(restos_2['latitude'] != 0) | (restos_2['longitude'] != 0)].copy()

# Reset index of the new DataFrame
restos_3.reset_index(drop=True, inplace=True)

# Now, 'restos_3' contains rows without 0 values in both "latitude" and "longitude" columns
print(restos_3)
```

```python
# Get the shape of the df
num_rows = restos_3.shape[0]

# Print the number of rows
print("Number of rows:", num_rows)

# Get the column names and data types
column_info = restos_3.dtypes

# Print the column names and data types
print("Column names and data types:")
for column_name, dtype in column_info.items():
    print(f"{column_name}: {dtype}")
```

```python
# Get the unique values in the "dba" column
unique_dbas = restos_3['dba'].unique()

# Count the number of unique values
num_unique_dbas = len(unique_dbas)

print("Number of unique values in the 'dba' column:", num_unique_dbas)

print("Unique values in the 'dba' column:")
for dba in unique_dbas:
    print(dba)
```

```python
# Get the unique values in the "violation_code" column
unique_vc = restos_3['violation_code'].unique()

# Count the number of unique values
num_unique_vc = len(unique_vc)

print("Number of unique values in the 'violation_code' column:", num_unique_vc)

print("Unique values in the 'violation_code' column:")
for vc in unique_vc:
    print(vc)
```

```python
# Get the unique values in the "camis" column
unique_camis = restos_3['camis'].unique()

# Count the number of unique values
num_unique_camis = len(unique_camis)

print("Number of unique values in the 'camis' column:", num_unique_camis)

print("Unique values in the 'camis' column:")
for camis in unique_camis:
    print(camis)
```

```python
# Write the dataframe "restos_3" to a CSV file
restos_3.to_csv("restos_3_mar_24.csv", index=False)
```
